```python
X1 = sm.add_constant(X)
```

```python
model = sm.OLS(y1,X1).fit()
```

```python
print(model.summary())
```

```python
coefficients = model.params
p_values = model.pvalues
```

```python
print(p_values < 0.05)   
```
