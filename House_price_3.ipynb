```python
import pandas as pd
import numpy as np
train = pd.read_csv('/content/train.csv')
test = pd.read_csv('/content/test.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
train = train.fillna('mean')
test = test.fillna('mean')
train.describe()
```

```python
test.describe()
```

```python
train = train.drop('Id', axis =1)
test = test.drop('Id', axis =1)
```

```python
train.describe()
```

```python
X = pd.concat([train, test])
```

```python
import pandas as pd

# Concatenate df1 and df2
df_combined = pd.concat([train, test], ignore_index=True)

# Identify categorical columns
categorical_columns = df_combined.select_dtypes(include=['object']).columns

# Perform one-hot encoding
df_encoded = pd.get_dummies(df_combined, columns=categorical_columns)

# Split the combined DataFrame back into df1 and df2
train_encoded = df_encoded.iloc[:len(train)]
test_encoded = df_encoded.iloc[len(train):]

# Now df1_encoded and df2_encoded contain numerical representations of categorical variables, and the encoding is consistent across df1 and df2
```

```python
y = train_encoded['SalePrice']
```

```python
X = train_encoded.drop('SalePrice', axis =1)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X,y)
```

```python
y_train_pred = lr.predict(X)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_train_pred, y)     ### behenchod
```
