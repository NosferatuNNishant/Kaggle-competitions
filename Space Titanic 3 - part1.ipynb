### Replacing categorical variables with numbers

```python
mapping = {'Earth': 1, 'Europa': 2, 'Mars': 3  , np.nan: 1}
train_['HomePlanet']  = train_['HomePlanet'].replace(mapping)       ## Homeplanet
test_['HomePlanet']  = test_['HomePlanet'].replace(mapping) 
```

```python
mapping = {True: 1, False: 0, np.nan: 0}       ## cryosleep
train_['CryoSleep']  = train_['CryoSleep'].replace(mapping)  
test_['CryoSleep']  = test_['CryoSleep'].replace(mapping)
```

```python
mapping = {'TRAPPIST-1e': 1, '55 Cancri e': 2, 'PSO J318.5-22': 3 ,np.nan: 1}       ## Destination
train_['Destination']  = train_['Destination'].replace(mapping) 
test_['Destination']  = test_['Destination'].replace(mapping) 
```

## let us try modelling now

```python
train1 = train_.drop('Cabin', axis = 1)
train1
```

```python
test1 = test_.drop('Cabin', axis = 1)
test1
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
logit = LogisticRegression()
```

```python
logit.fit(train1, y1)
```

```python
y_train_pred = logit.predict(train1)
print(accuracy_score(y_train_pred, y1))         ### 78.8% accuracy with Logistic Regression
```

```python
y_test_pred_lr = logit.predict(test1)
print(y_test_pred_lr)
```
