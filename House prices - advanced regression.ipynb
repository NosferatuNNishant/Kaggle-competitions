## Approach 1: Using only the continuous variables and checking their regression coefficients

```python
train['SalePrice'].dtype
```

```python
train1 = train.select_dtypes(include = ['float64', 'int64'])
test1 = test.select_dtypes(include = ['float64', 'int64'])
```

```python
len(train1.columns)
```

#### I will check which variables matter and which are useless data.

```python
from sklearn.linear_model import LinearRegression
```

```python
train1.columns
```

```python
train2 = train1.dropna()
```

```python
X = train2.drop('SalePrice', axis =1)
y = train2['SalePrice']
```

```python
X.head(2)
```

```python
y.head(2)
```

```python
X.describe()
```

```python
model = LinearRegression()
model.fit(X,y)
```

##### Now checking for the variable coefficients

```python
coefficients = model.coef_
features = X.columns

print("Regression Coefficients:\n")
for i, name in enumerate(features):
    print(f"- {name}: {coefficients[i]:.4f}")  # Format coefficients with 4 decimal places
```

### checking with p value - less than 0.05 is fine

```python
import statsmodels.api as sm
model = sm.OLS(y,X).fit()
```

```python
print(model.summary())
```

```python
coefficients = model.params
p_values = model.pvalues
## significant coefficients
sig_coefs = [(name, coef) for name, coef, p_value in zip(X.columns, coefficients, p_values) if p_value <=0.05]
```

```python
sig_coefs
```

```python
cols = ["Variable", "Coefficient"]
```

```python
VC = pd.DataFrame(sig_coefs, columns = cols)     ### VC for variable and coefficient
```

```python
VC
```

```python
list1 = VC['Variable']
print(list1)
```

```python
X2 = X[list1]
```

```python
X2.head()
```

### Now this X2 has only the numerical variables that matter - let's use this to make a regression model

```python
model = LinearRegression()
model.fit(X2, y)
```

```python
test.head(2) ### this also needs to be changed
```

```python
test2 = test[list1]
```

```python
test2.describe() ### missing value imputation required
```

```python
test3 = test2.fillna(test2.mean())    ### test 3 is the data that I am gonna use
```

```python
y_pred = model.predict(test3)
```

```python
from sklearn.metrics import r2_score         ### r2
from sklearn.metrics import mean_absolute_error   #### MAE
```

```python
#### Oh wait, what do I test it against? 
y_train_pred = model.predict(X2)
```

```python
mean_absolute_error(y_train_pred, y)
```

```python
r2_score(y_train_pred, y)
```
